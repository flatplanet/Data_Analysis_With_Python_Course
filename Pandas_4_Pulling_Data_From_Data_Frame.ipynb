```python
# Describe a specific column - Breed
my_df["Breed"].describe()
```

```python
# Describe a specific column - Color
my_df['Color'].describe()
```

```python
# Describe a specific column - DogName
my_df['DogName'].describe()
```

```python
# Count Data In Dataframe - Count Distinct Values - Descending
my_df["Color"].value_counts()
```

```python
# Count Distinct Values - Ascending
my_df["Color"].value_counts(ascending=True)
```

```python
# Dog Names - without NaN
my_df["DogName"].value_counts()
```

```python
# Dog Names - with NaN
my_df["DogName"].value_counts(dropna=False)
```

```python
# Get relative Frequency - Percentage
my_df["Color"].value_counts(normalize=True)
```

```python
# Get Specific Item Count (Get all white dogs)
print(my_df["Color"].value_counts()["WHITE"])
```

```python
# Count Unique Values - Size
my_df.groupby('Color').size()
```

```python
# Count Unique Values - Count
my_df.groupby("Color").count()
```
