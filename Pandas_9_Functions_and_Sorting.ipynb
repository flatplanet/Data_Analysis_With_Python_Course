# Pandas - Functions and Sorting

```python
import pandas as pd
import numpy as np
```

```python
# Create Dataframe
stuff = {
    'Corporation':["Apple", "Google", "Meta", "Apple", "Google", "Meta"],
    'Employees':['John', 'April', 'Wes', 'Aspen', 'Beth', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]}
my_df = pd.DataFrame(stuff)
my_df
```

```python
 # Create Function to change $ to thousands
 def times1000(x):
  return x * 1000
```

```python
# Apply the function
my_df['Salary'].apply(times1000)
```

```python
# Format it nicer with decimals
def times1000(x):
  return format(x * 1000, ",d")
```

```python
my_df['Salary'].apply(times1000)
```

```python
# Make it into a dataframe
pd.DataFrame(my_df['Salary'].apply(times1000))
```

```python
# Add the new data to the old dataframe
my_df["Salary"] = my_df['Salary'].apply(times1000)
my_df
```

```python
# Add last name to names
def namer(x):
  if x == "John":
    return "John Elder"
  elif x == "April":
    return "April Elder"
  else:
    return x
```

```python
# Apply it
my_df["Employees"].apply(namer)
```

```python
# Use lambdas inline
my_df["Salary"].apply(lambda x: format(x * 1000, ',d'))
```

```python
# Make dataframe
pd.DataFrame(my_df["Salary"].apply(lambda x: format(x * 1000, ',d')))
```

```python
# Append that stuff back to original dataframe
my_df["Salary"] = my_df["Salary"].apply(lambda x: format(x * 1000, ',d'))
```

```python
my_df
```

```python
# Apply Functions to multiple columns (corp and emps)
def names(x):
  return f"Codemy: {x}"
```

```python
# Apply To Two Columns
my_df[["Corporation", "Employees"]].apply(names)
```

```python
# Append to Current DataFrame
pd.DataFrame(my_df[["Corporation", "Employees"]].apply(names))
```

```python
my_df
```

```python
# Sort Salaries from lowest to highest Ascending
my_df.sort_values("Salary")
```

```python
# Sort Salaries from Highest to Lowest Descending
my_df.sort_values("Salary", ascending=False, inplace=True)
```

```python
# Not Permanent without inplace=True
my_df
```

```python
# Sort Strings Alphabetically Ascending A to Z
my_df.sort_values("Corporation")
```

```python
# Sort Strings Alphabetically Descending Z to A
my_df.sort_values("Corporation", ascending=False)
```
