# Charts and Graphs with Matplotlib


```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create Random Number DataFrame
my_df = pd.DataFrame(randn(100, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Histogram
my_df["Wed"].hist()
```

```python
# Histogram with smaller bins, 50, 20
my_df["Wed"].hist(bins=20)
```

```python
# Second Method to create Histograms
my_df["Wed"].plot(kind="hist")
```

```python
# Original method without gridlines
my_df["Wed"].hist(bins=20, grid=False)
```

```python
# Add a Legend
my_df["Wed"].hist(bins=20, grid=False, legend=True)
```

```python
# Area Plots
# Create Random Number DataFrame
my_df = pd.DataFrame(randn(20, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Create Area Plot
my_df.plot(kind="area", stacked=False)
```

```python
# Absolute Values
my_df.abs().plot(kind="area")
```

```python
# Shading
my_df.abs().plot(kind="area", alpha=0.3) # Closer to 0 is more transparent
```

```python
# Title
my_df.abs().plot(kind="area", alpha=0.3, title="My Very Cool Area Plot")
```

```python
# Remove Legend
my_df.abs().plot(kind="area", alpha=0.3, title="My Very Cool Area Plot", legend=False)
```

```python
# Plot only one column
my_df['Mon'].abs().plot(kind="area", alpha=0.3, title="My Very Cool Area Plot")
```

```python
# Plot 2 Columns
my_df[["Mon", "Wed"]].abs().plot(kind="area", alpha=0.3, title="My Very Cool Area Plot", legend=False)
```

```python
# Get info on a Function Shift Tab on the function - remove abs
my_df.plot(kind="area")
```

```python
# Add Table
my_df.abs().plot(kind="area", alpha=0.3, title="My Very Cool Area Plot", legend=False, table=True)
```
