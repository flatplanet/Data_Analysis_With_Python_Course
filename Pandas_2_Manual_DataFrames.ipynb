# Pandas - Manual DataFrames

```python
# Import Pandas
import pandas as pd
import numpy as np
from numpy.random import rand
```

```python
# Create a Random Data
my_data = rand(4,3) # rows/columns
print(my_data)
```

```python
# Create headers
my_rows = ["A", "B", "C", "D"]
my_cols = ["Monday", "Tuesday", "Friday"]
```

```python
# Create our DataFrame!!
my_df = pd.DataFrame(my_data, my_rows, my_cols)
```

```python
# print
print(my_df)
```
