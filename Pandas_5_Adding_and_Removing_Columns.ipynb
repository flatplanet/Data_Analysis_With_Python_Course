# Pandas - Adding and Removing

```python
import pandas as pd
import numpy as np
# Import CSV File
my_df = pd.read_csv('/content/dog_data_short.csv')
```

```python
my_df
```

```python
# Add Column from List (must match)
gender_list = ["Male", "Female", "Male", "Male", "Female"]
# Create a column and assign the stuff
my_df["Gender"] = gender_list
my_df
```

```python
# Add Default Values
my_df["Alive/Dead"] = [True]*len(my_df)
my_df
```

```python
# Add Null Values NaN
my_df["ShowDogs"] = [np.nan]*len(my_df)
my_df
```

```python
# Add Column with .insert() (allows you to specify where)
my_df.insert(1,"Adopted", [True]*len(my_df), True)
my_df
```

```python
# Add Column with .assign() (creates a new DataFrame)
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```

```python
my_df
```

```python
# Remove Column
my_df.drop("ShowDogs")
```

```python
# Axis is column heading or row headings (0=rows, 1=columns)
my_df.drop("ShowDogs", axis=1, inplace=True)
```

```python
# run again my_df and gender is back...inplace=False default
my_df
```

```python
# run one more time just to see...
my_df
```

```python
# Remove Rows
my_df.drop(3, axis=0, inplace=False)
```
