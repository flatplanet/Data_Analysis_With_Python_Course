# Seaborn Charts

```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
```

```python
my_df = pd.DataFrame(randn(100, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Basic histogram
sns.histplot(data=my_df, x="Mon")
```

```python
# Histogram with customization
sns.histplot(data=my_df, x="Mon", bins=15, color="green")
```

```python
# Histogram with KDE overlay
sns.histplot(data=my_df, x="Mon", kde=True, bins=15, color="coral")
```

```python
# Histogram with transparency and different stat
sns.histplot(data=my_df, x="Mon", kde=True, bins=15, color="coral", alpha=0.3, stat="density")
```

```python
# Basic bar chart (using mean values)
means = my_df.abs().mean()
days = my_df.columns
sns.barplot(x=days, y=means)
```

```python
# Bar chart with colors
sns.barplot(x=days, y=means, palette="coolwarm", hue=days)
```

```python
# Horizontal bar chart
sns.barplot(y=days, x=means, palette="Set2", hue=days)
```

```python
# Bar chart with transparency and width
sns.barplot(x=days, y=means, palette="coolwarm", hue=days, width=0.2, alpha=0.3)
```

```python
# Basic line chart
sns.lineplot(data=abs(my_df), x=my_df.index, y="Mon")
```

```python
# Multiple lines line chart
sns.lineplot(abs(my_df), alpha=0.9)
```

```python
# Line chart with markers and styling
sns.lineplot(data=abs(my_df), x=my_df.index, y="Mon", marker="o", linewidth=2, color="purple")
```

```python
# Line chart with transparency
sns.lineplot(data=abs(my_df), x=my_df.index, y="Mon", marker="o", linewidth=2, color="purple", alpha=0.3)
```

```python
# Basic scatterplot
sns.scatterplot(data=my_df, x="Mon", y="Tues")
```

```python
# Scatterplot with color and size
sns.scatterplot(data=my_df, x="Mon", y="Tues", color="red", s=200)
```

```python
# Scatterplot with hue (color by values)
sns.scatterplot(data=my_df, x="Mon", y="Tues", hue=my_df["Mon"], palette="viridis", s=200)
```

```python
# Scatterplot with size based on values and transparency
sns.scatterplot(data=my_df, x="Mon", y="Tues", hue="Mon", palette="viridis", alpha=0.3, sizes=(20, 800), size=abs(my_df["Tues"]))
```

```python
# Basic boxplot
sns.boxplot(data=my_df)
```

```python
# Boxplot with colors
sns.boxplot(data=my_df, palette="Set2")
```

```python
# Horizontal boxplot
sns.boxplot(data=my_df, orient="h", palette="Set2")
```

```python
# Boxplot with custom width and without outliers
sns.boxplot(data=my_df, palette="Set2", showfliers=False)
```

```python
# Basic KDE plot
sns.kdeplot(data=my_df, x="Mon", fill=True)
```

```python
# Multiple KDE plots
sns.kdeplot(data=my_df, x="Mon", fill=True, label="Mon")
sns.kdeplot(data=my_df, x="Thur", fill=True, label="Tues")
```

```python
# KDE plot with bandwidth adjustment (smoother)
sns.kdeplot(data=my_df, x="Mon", fill=True, label="Mon", bw_adjust=2, alpha=0.3)
```

```python
# KDE plot with different bandwidth (rougher) and linewidth
sns.kdeplot(data=my_df, x="Mon", fill=True, label="Mon", bw_adjust=0.5, alpha=0.3, linewidth=3, color="purple")
```
